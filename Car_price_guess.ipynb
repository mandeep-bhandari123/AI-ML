```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

To find the price of car according to data set

```python
x_train = np.array([172 , 340])
y_train = np.array([28107 , 42137])
```

```python
m = len(x_train)#Number of datas in training set
print(f"Number of datas in training set is : {m}")
```

```python
i = 0 
x_i = x_train[i]
y_i = y_train[i]   
print(f"(x^{i}, y^{i}) = ({x_i}, {y_i})")
```

```python
#Poting the data 
plt.scatter(x_train, y_train , marker = "*" , c = "w")
plt.title("Car price prediction")
plt.ylabel('Price of car in $')
plt.xlabel("Horse power of car")
plt.show()
```

```python

#Let w be the slope and b be the y-intercept of the line 
w = (y_train[1]-y_train[0])/(x_train[1]-x_train[0]) #Slope formula (y2-y1)/(x2-x1)
b = 17174.047495 #y-intercept 
print(f"w:{w}")
print(f"b:{b}")
```

```python
def calculate_model_output( w , b , x):
    m = x.shape
    f_wb = np.zeros(m)
    for i in range (len(x)):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wv =calculate_model_output(w, b, x_train)
#plot our model prediction
plt.plot(x_train , tmp_f_wv , c = "r" , label = "Our prediction")
plt.scatter(x_train, y_train , marker = "*" , c = "w" , label = "Actual Values")
plt.title("Car price prediction")
plt.ylabel('Price of car in $')
plt.xlabel("Horse power of car")
plt.legend()
plt.show()
```
